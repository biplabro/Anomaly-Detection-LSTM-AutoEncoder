```python
# generate a sine wave & obtain the timeseries values
# y(t) = A sin(ωt + Φ)

# curve for both positive and negative half
```

```python
# importing libraries first
import numpy as np
import matplotlib.pyplot as plt
```

```python
# amplitude A
# positive half
a = 10

# negative half
# amplitude can never be negative but to obtain negative y values, - is considered in the equation
# yN = -(a*np.sin(2*np.pi*F*time + phi))

# angular frequency ω = 2.Pi.F
# need to fit 200 data points, in 100 seconds, half a sine period
# F = (1/T); period is 200 seconds; we need half of the cycle, thus range 100 is selected np.arange(0,100,0.5)
# F= 1/200 = 0.005
F = 0.005

# time t
# T = 0

# phase Φ
phi = 0

# sampling rate (number of samples/data points per second)
# sr = 100
```

```python
# Generate the x and y axis data
# yP = pisitive half

time = np.arange(0,100,0.5) # (start, end, interval)

yP = a*np.sin(2*np.pi*F*time + phi)
```

```python
# plot the graph

# plot window
# plt.figure(figsize=(10,4))

# input parameters here
plt.plot(time,yP) 
plt.title("Sine Wave")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
yP
```

```python
# add an offset of 70
yP=yP+70
```

```python
yP
```

```python
# plot upscaled data
# input parameters here
plt.plot(time,yP) 
plt.title("Sine Wave")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Generate the x and y axis data
# yN = Negative half

time = np.arange(0,100,0.5) # (start, end, interval)
# sin(Pi+x)= −sin(x)
# Negative co-efficient for obtaining the datapoints in the -Y axis
yN = -(a*np.sin(2*np.pi*F*time + phi))
```

```python
# plot the graph
# input parameters here
plt.plot(time,yN) 
plt.title("Sine Wave")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
yN
```

```python
# add an offset of 70
# +offset, since amplitude is already negative
yN=yN+70
```

```python
yN
```

```python
# plot downscaled data
# input parameters here
plt.plot(time,yN) 
plt.title("Sine Wave")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```
